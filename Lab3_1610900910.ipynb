```

import pandas as pd
import io

io.StringIO(uploaded["datasets_448905_849030_akpam.csv"].decode('utf_8'))
df = pd.read_csv(io.StringIO(uploaded["datasets_448905_849030_akpam.csv"].decode('utf_8')))
```

 -code สร้าง Bar Chart 

```
from matplotlib import pyplot as plt
plt.bar(df.Akpam, df.ips) #plot (Line Chart) --> bar (Bar Chart) --> scatter (Scatter Plot)
plt.title("Akpam vs IPS")
plt.xlabel("Akpam")
plt.ylabel("ips")
plt.rcParams["figure.figsize"] = (20,10)
data_line = plt.xticks(rotation=90)
```

```
file_name = "datasets_48230_87684_btcusd.csv"
data = uploaded[file_name].decode('utf_8').split("\r\n")
```

```
for i in range(len(data)):
  data[i] = data[i].split(",")

print(data)
```

```

import pandas as pd
import io

io.StringIO(uploaded["datasets_48230_87684_btcusd.csv"].decode('utf_8'))
df = pd.read_csv(io.StringIO(uploaded["datasets_48230_87684_btcusd.csv"].decode('utf_8')))
```

```
print(df)
```

```
#Ploting Line Plots 
from matplotlib import pyplot as plt
plt.plot(df.ds, df.y) #plot (Line Chart) --> bar (Bar Chart) --> scatter (Scatter Plot)
plt.title("BTC (USD) data between 2012-2018")
plt.xlabel("date")
plt.ylabel("price")
plt.rcParams["figure.figsize"] = (20,10)
data_line = plt.xticks(rotation=90)
```
